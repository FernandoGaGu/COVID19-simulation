```python
import random
random.seed(7777)
# Import simamics module componentes
from simamics.dynamic import Interaction
from simamics.population import Population
from simamics.pathogen import Virus
from simamics.dynamic import Interaction, simulate
```

```python
# Function that defines the number of interactions based on population size, 
# movement and space
interaction = Interaction(lambda size, mov, space: int((size*(pow((2*mov), 2)))/space))
interaction.plotFunction(populationSpace=400)
```

```python
# Population with high activity and low number of diagnosis test
population = Population(initialCases=10, populationSize=4000, space=3000)

virus = Virus(
    lifeTime=10, deathRate=0.01, lightPercentage=0.35, mildPercentage=0.35, severePercentage=0.2,
    asymptomaticPercentage=0.1, transmissionPercentage=0.5)

report = simulate(population=population, virus=virus, iterations=50, populationActivity=0.8,
                  interaction=interaction, diagnosisPercentage=0.2)

report.backend = "matplotlib"
report.plotStatistics(title="Coronavirus evolution")
```

```python
# Population with high activity and high number of diagnosis test
population = Population(initialCases=10, populationSize=4000, space=3000)

virus = Virus(
    lifeTime=10, deathRate=0.01, lightPercentage=0.35, mildPercentage=0.35, severePercentage=0.2,
    asymptomaticPercentage=0.1, transmissionPercentage=0.5)

report = simulate(population=population, virus=virus, iterations=100, populationActivity=0.8,
                  interaction=interaction, diagnosisPercentage=0.8)

report.backend = "matplotlib"
report.plotStatistics(title="Coronavirus evolution")
```
