```python
import random
random.seed(7777)

# Import simamics module componentes
from simamics.dynamic import Interaction
from simamics.population import Population
from simamics.pathogen import Virus
from simamics.dynamic import Interaction, simulate
```

```python
# Function that defines the number of interactions based on population size, 
# movement and space.
interaction = Interaction(
    lambda size, mov, space: int((size*(pow((2*mov), 2)))/space)
)

interaction.plotFunction(populationSpace=200)
```

# Normal activity + Low number of test

```python
#Population with normal activity and a low number of tests
population = Population(initialCases=10, populationSize=4000, space=2000)

virus = Virus(
    lifeTime=12, deathRate=0.01, lightPercentage=0.35, mildPercentage=0.35, severePercentage=0.2,
    asymptomaticPercentage=0.1, transmissionPercentage=0.5)

report = simulate(population=population, virus=virus, iterations=50, populationActivity=0.8,
                  interaction=interaction, diagnosisPercentage=0.2)

report.plotStatistics(title="Coronavirus evolution")
```

```python
report.plotDeaths(title="Coronavirus deaths")
```
